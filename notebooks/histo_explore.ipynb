```python
import os

histo = [0]*5000

hd = os.path.join(os.getcwd(), 'histo')

for i in range(50):
    with open(os.path.join(hd, 'histo'+str(i)+'.txt')) as f:
        h = f.read()
        h = h[1:-1]
        h = h.split(', ')
        h = [int(x) for x in h]
        for j in range(5000):
            histo[j] += h[j]

# cut off all the extra 0s in tail
print(sum(histo))
print(sum(histo[500:]))
histo = histo[:500]
print(histo)
print(histo[476])
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

fig, ax = plt.subplots(figsize=(13,12))

x = []
y = []
for i,v in enumerate(histo[:31]):
    x.append(i)
    y.append(v)

#######################
#VERTICAL BAR CHART
#######################
bars = plt.bar(x, y) #,width=5
#ax = plt.gca()
#ax.tick_params(axis='x', labelrotation=45)
# xlocs, xlabs = plt.xticks()
# for i, v in enumerate(y):
#     plt.text(xlocs[i], v+100, str(v))
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x()+.5, yval+100, yval, horizontalalignment="center")

#######################
# HORIZONTAL BAR CHART
#######################
# #x.reverse()
# y.reverse()
# for i,v in enumerate(y):
#     ax.text(v+70, i-.2, str(v)) #, color='blue', fontweight='bold')
# plt.barh(x, y)
# #plt.xticks(rotation=90)
# plt.yticks(range(30), range(29,-1,-1))#, rotation=45)

plt.title("\nChord Progression Uniqueness\n", fontsize=30)
plt.xlabel('\nNumber of Songs Returned\n', fontsize=20)
plt.ylabel('\nAPI Calls\n(Chord Progressions)\n', fontsize=20)
plt.show()

# JESSE ADD THE NUMBER OF FAILED CALLS TO histo[0] SO YOU CAN VIEW THE FAILURES
```
