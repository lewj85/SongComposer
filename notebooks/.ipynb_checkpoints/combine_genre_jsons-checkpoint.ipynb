```python
import json
```

```python
big_dict = {}

for i in range(50):
    a = []
    with open('../genres/newprobs' + str(i) + 'genre.json') as f:
        a = json.load(f)
    for key,value in a.items():
        for song in range(len(a[key])):
            if not a[key][song]['genre']:
                a[key][song]['genre'] = ['Unknown']
    big_dict.update(a)
```

```python
big_dict['1,5,6,4']
```

```python
#unique = { each['Name'] : each for each in big_dict }.values()

clean_dict = {}
for eachchord in big_dict.keys():
    clean_dict[eachchord] = []
    for eachsong in big_dict[eachchord]:
        if eachsong not in clean_dict[eachchord]:
            clean_dict[eachchord].append(eachsong)
```

```python
clean_dict['1,5,6,4']
```

```python
with open('genres_combined.json', 'w') as outfile:
    json.dump(clean_dict, outfile)
```
