```python
import json

with open('url.json', 'r') as f:
    urls = json.load(f)

genres = {}

for u in urls:
    for i,g in enumerate(urls[u]):
        if genres.get(g, False):
            genres[g] += 1
        else:
            genres[g] = 1

#print(genres)
genres['Childrens'] = genres["Children's"]
genres.pop("Children's")
#print(genres)

genres = sorted(genres.items(), key=lambda item: item[1], reverse=False)
#print(genres)
genres = {k: v for k, v in genres}
print(genres)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

fig, ax = plt.subplots(figsize=(12,14))

keys = list(genres.keys())
values = list(genres.values())

#######################
# VERTICAL BAR CHART
#######################
# plt.bar(keys, values)
# ax = plt.gca()
# ax.tick_params(axis='x', labelrotation=45)
# plt.show()

#######################
# HORIZONTAL BAR CHART
#######################
for i, v in enumerate(values):
    ax.text(v+10, i-.2, str(v)) #, color='blue', fontweight='bold')
plt.barh(keys, values)
#plt.xticks(rotation=90)

plt.title("\nSongs per Genre\n", fontsize=30)
plt.xlabel('\nSongs\n', fontsize=20)
plt.ylabel('\nGenre\n', fontsize=20)
plt.show()
```
