```python
import os

with open(os.path.join(os.getcwd(), 'speed.txt')) as f:
    s = f.read()
    speeds = [float(x) for x in s.split()]

minn = min(speeds)
maxx = max(speeds)

print('Min', minn)
print(minn//60., 'minutes,', minn%60., 'seconds\n')
print('Max', maxx)
print(maxx//60., 'minutes,', maxx%60., 'seconds\n')

print(speeds)
```

```python
from matplotlib import pyplot as plt
%matplotlib inline

plt.plot(speeds)

plt.title("\nSpeed of API Scripts\n", fontsize=30)
plt.xlabel('\nBatch of 500 Chord Progressions\n', fontsize=20)
plt.ylabel('\nSeconds\n', fontsize=20)

plt.show()
```
