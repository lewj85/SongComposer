```python
directory = os.getcwd()
jsonfile = 'chordProbs3new.json'

with open(os.path.join(directory, '..', 'data', jsonfile)) as f:
    data = json.load(f)

for d in data:
    print(d)
    break
```

```python
chord1 = []
chord2 = []
chord3 = []
chord4 = []
probabilities = []

for d in data:
    chord1.append(d['child_path'].split(',')[0])
    chord2.append(d['child_path'].split(',')[1])
    chord3.append(d['child_path'].split(',')[2])
    chord4.append(d['child_path'].split(',')[3])
    #first_3_chords.append(d['child_path'].split(',')[:-1])
    #fourth_chords.append(d['chord_ID'])
    probabilities.append(d['probability'])

df = pd.DataFrame(list(zip(chord1, chord2, chord3, chord4, probabilities)),
               columns =['Chord1', 'Chord2', 'Chord3', 'Chord4', 'Probability'])
df.head()
```

```python
chord_dict1 = {}
chord_dict2 = {}
chord_dict3 = {}
chord_dict4 = {}
for i,u in enumerate(np.unique(df['Chord1'])):
    chord_dict1[u] = i
for i,u in enumerate(np.unique(df['Chord2'])):
    chord_dict2[u] = i
for i,u in enumerate(np.unique(df['Chord3'])):
    chord_dict3[u] = i
for i,u in enumerate(np.unique(df['Chord4'])):
    chord_dict4[u] = i

print(chord_dict1)
```

```python
X = pd.DataFrame(columns=['1_'+str(x) for x in chord_dict1.keys()] + \
                         ['2_'+str(x) for x in chord_dict2.keys()] + \
                         ['3_'+str(x) for x in chord_dict3.keys()])
X.head()
```

```python
# Prepare a dictionary of estimators after instantiating each one of them
ESTIMATORS = {
    "Extra trees": ExtraTreesRegressor(n_estimators=10,
                                       max_features=32,     # Out of 20000
                                       random_state=0),
    "K-NN": KNeighborsRegressor(),                          # Accept default parameters
    "Linear regression": LinearRegression(),
    "Ridge": RidgeCV(),
    "Lasso": Lasso(),
    "ElasticNet": ElasticNet(random_state=0),
    "RandomForestRegressor": RandomForestRegressor(max_depth=4, random_state=2),
    "Decision Tree Regressor":DecisionTreeRegressor(max_depth=5),
    "MultiO/P GBR" :MultiOutputRegressor(GradientBoostingRegressor(n_estimators=5)),
    "MultiO/P AdaB" :MultiOutputRegressor(AdaBoostRegressor(n_estimators=5))
}
```

```python
# 9.1 Create an empty dictionary to collect prediction values
y_test_predict = dict()
y_mse = dict()

for name, estimator in ESTIMATORS.items():     
    estimator.fit(X_train, y_train)                    # fit() with instantiated object
    y_test_predict[name] = estimator.predict(X_test)   # Make predictions and save it in dict under key: name
    y_mse[name] = mean_squared_error(y_test, y_test_predict[name])
    plt.figure(figsize=(8,6))
    plt.scatter(y_test_predict[name], y_test, cmap='plasma')
    plt.title(name, fontsize=24)
    plt.xlabel('Prediction', fontsize=16)
    plt.ylabel('Label', fontsize=16)
    plt.show()
    print('RMSE for ', name, ' is ', y_mse[name], '\n\n\n\n')
```

```python
# design model
model = Sequential()
model.add(Dense(128, input_dim=(143), activation="linear"))
model.add(Dense(64, activation="linear"))
model.add(Dense(32, activation="linear"))
model.add(Dense(51))
model.compile(loss="mse", optimizer="adam")
 
model.summary()
```

```python
# Train the model, iterating on the data in batches of 4 samples
model.fit(X_train, y_train, epochs=5, batch_size=8)
```

```python
# Now train on the whole X, y and save model to .h5 file

model.compile(loss="mse", optimizer=SGD(lr=0.01))

def scheduler(epoch, lr):
    if epoch > 0:
        lr /= 2
    print('lr', lr)
    return lr

#round(model.optimizer.lr.numpy(), 5)
#0.01

callback = LearningRateScheduler(scheduler)
history = model.fit(X, y,
                    batch_size=4,
                    epochs=10,
                    callbacks=[callback],
                    verbose=1)

model.evaluate(X, y)
```
